```python
with open('si.txt', 'r') as file:
    lines = file.readlines()
b=[float(i[:-1]) for i in lines]
print("Максимальная температура:",max(b))
print("Минимальная температура:",min(b))
print("Средняя температура:",round(sum(b)/len(b),2))  
```

```python
import matplotlib.pyplot as plt
y=b
x=range(len(b))
plt.plot(x,y)  
plt.show()
```
