```python
class Player:
    def __init__(self, new_nickname='Игрок без никнейма'):        
        self._nickname = new_nickname 
        print(f'подключился новый игрок {self._nickname}')
        self._power=0
        self._hp=100
        self._x=0
        self._y=0
        self._a=1
        self._b=0
        
    def set_nickname(self, new_nickname):
        self._nickname = new_nickname 
    def get_nickname(self):
        return self._nickname
    def collect(self):
        self._power+=1
        print(f"игрок с никнеймом {self._nickname} собирает предметы, его сила возросла до {self._power}")
    def set_up(self):
        self._x=self._x
        self._y=self._y+self._power*self._a
        print(f"Игрок с никнеймом {self._nickname} переместился в поле ({self._x},{self._y})")
    def set_down(self):  
        self._x=self._x
        self._y=self._y-self._power*self._a
        print(f"Игрок с никнеймом {self._nickname} переместился в поле ({self._x},{self._y})")
    def set_right(self):
        self._x=self._x+self._power*self._a
        self._y=self._y
        print(f"Игрок с никнеймом {self._nickname} переместился в поле ({self._x},{self._y})")
        if  self._x>50:
            print(f"игрок с никнеймом {self._nickname} победил!!!")
    def set_left(self):
        self._x=self._x-self._power*self._a
        self._y=self._y
        print(f"Игрок с никнеймом {self._nickname} переместился в поле ({self._x},{self._y})")
    def wing(self):
        self._a=self._a+1
        print(f"Игрок с никнеймом {self._nickname} нашел крыло,теперь у него {self._a} крыла")
    def radiation(self):
        self._hp=self._hp-20
        print(f"{self._nickname} попал в радиацию,теперь у него {self._hp} здоровья")
        if 20<self._hp<100:
            print(f" игроку с никнеймом {self._nickname} нужна аптечка")
        if self._hp==20:
            print(f"игрок с никнеймом {self._nickname}  на грани гибели")
    def medication(self):
        self._hp=self._hp+20
        self._b=self._b-1
        print(f"игрок с никнеймом {self._nickname} использовал аптечку,теперь его здоровье {self._hp},теперь у него {self._b} аптечек")
    def medicine(self):
        self._b=self._b+1
        print(f"игрок с никнеймом {self._nickname} нашел аптечку,теперь у него {self._b} аптечек")
    
p=Player("Di")
p.collect()
p.wing()
p.medicine()
p.set_up()
p.set_right()
p.radiation()
p.medication()
p.collect()
p.set_right()
p.set_right()
p.radiation()
p.radiation()
p.radiation()
p.set_up()
p.wing()
p.set_up()
p.radiation()
p.set_up()
p.collect()
p.collect()
p.medicine()
p.medication()
p.set_right()
p.set_right()
p.radiation()
p.wing()
p.collect()
p.set_right()
    
```
